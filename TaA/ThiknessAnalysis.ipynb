# Thickness Analysis Test
## Analytical aproximation

```python
import matplotlib.pyplot as plt

# Intervalos de energía (keV)
energy_intervals = [
    (0.0061305, 0.01),
    (0.01, 0.05),
    (0.05, 0.1),
    (0.1, 0.9),
    (0.9, 3.5),
    (3.5, 6),
    (6, 20),
    (20, 25),
    (25, 100),
    (100, 200),
    (200, 300),
    (300, 432.9451)
]

# Valores de thickness correspondientes (en nm, mm ya convertidos a nm)
thickness_values = [
    60,          # para 0.0061305 - 0.01 keV
    80,          # para 0.01 - 0.05 keV
    120,         # para 0.05 - 0.1 keV
    280,         # para 0.1 - 0.9 keV
    700,         # para 0.9 - 3.5 keV
    2200,        # para 3.5 - 6 keV
    9000,        # para 6 - 20 keV
    40000,       # para 20 - 25 keV
    0.1 * 1e6,   # para 25 - 100 keV
    1 * 1e6,     # para 100 - 200 keV
    5 * 1e6,     # para 200 - 300 keV
    13 * 1e6     # para 300 - 432.9451 keV
]

# Cálculo del punto medio de cada intervalo
energy_midpoints = [(min_e + max_e) / 2 for min_e, max_e in energy_intervals]

# Gráfica
plt.plot(energy_midpoints, thickness_values, marker='o', linestyle='-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Energy (keV)')
plt.ylabel('Thickness (nm)')
plt.title('Thickness vs Energy Midpoints')
plt.grid(True)
plt.show()
```
